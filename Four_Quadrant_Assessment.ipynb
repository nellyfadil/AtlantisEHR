```python
import matplotlib.pyplot as plt

def calculate_triangle_area(x1, y1, x2, y2, x3, y3):
    area = 0.5 * abs(x1 * (y2 - y3) + x2 * (y3 - y1) + x3 * (y1 - y2))
    return area

def plot_points_and_triangle(x1, y1, x2, y2, x3, y3):
    # Plot points
    plt.scatter([x1, x2, x3], [y1, y2, y3], color='red', label='Vertices')

    # Plot triangle
    plt.plot([x1, x2, x3, x1], [y1, y2, y3, y1], color='blue', label='Triangle')

    # Set axis limits
    plt.xlim(-15, 15)
    plt.ylim(-15, 15)

    # Add labels and legend
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
    plt.legend()

    # Show the plot
    plt.show()

# Example coordinates of triangle vertices
x1, y1 = 1, 2
x2, y2 = 4, 6
x3, y3 = -2, 3

# Calculate and print the area of the triangle
triangle_area = calculate_triangle_area(x1, y1, x2, y2, x3, y3)
print(f"The overall area of the triangle is: {triangle_area}")

# Calculate the area of the triangle in quadrant one
quadrant_one_area = calculate_triangle_area(0, 0, x2, y2, x3, y3)
print(f"The area of the triangle in quadrant one is: {quadrant_one_area}")

# Plot points and triangle
plot_points_and_triangle(x1, y1, x2, y2, x3, y3)
```

```python
import matplotlib.pyplot as plt

def calculate_triangle_area(x1, y1, x2, y2, x3, y3):
    return 0.5 * abs(x1 * (y2 - y3) + x2 * (y3 - y1) + x3 * (y1 - y2))

# Vertices of the original triangle
x1, y1 = 1, 2
x2, y2 = 4, 6
x3, y3 = -2, 3

# Extra vertex in quadrant one
extra_x, extra_y = 5, 3

# Calculate overall area of the original triangle
triangle_area = calculate_triangle_area(x1, y1, x2, y2, x3, y3)

# Calculate the area of the portion of the triangle in quadrant one
quadrant_one_area = calculate_triangle_area(0, 0, x2, y2, x3, y3)

# Plotting
plt.figure(figsize=(8, 8))

# Plot original triangle
plt.plot([x1, x2, x3, x1], [y1, y2, y3, y1], label='Original Triangle', marker='o')

# Plot extra vertex in quadrant one
plt.plot(extra_x, extra_y, 'ro', label='Extra Vertex in Q1')

# Plot the triangle connecting vertices from quadrant two and three to the point in quadrant one
plt.plot([x2, x3, extra_x, x2], [y2, y3, extra_y, y2], label='New Triangle', linestyle='--', marker='o')

# Set axis limits
plt.xlim(-15, 15)
plt.ylim(-15, 15)

# Display labels and legend
plt.title('Triangles on Four-Quadrant Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

# Display the plot
plt.show()

# Print area information
print(f"Overall area of the original triangle: {triangle_area}")
print(f"Area of the portion of the triangle in quadrant one: {quadrant_one_area}")

```
